````python
def credit_card_payment(interest, payment ) :    
    bill = 200.00
    total = 0.0
    # calculate salaries for each year based on the given increase
    balance = [bill]
    ind = 1
    while balance[ind-1] > 0 and ind < 120 :
        if balance[ind-1] >= payment:
            balance.append( (balance[ind-1] - payment) * (1+interest) ) 
            total += payment 
        else : 
            balance.append( 0 )
            total  += balance[ind-1]
        ind += 1 
    balance.append(0)   
    # values for horizontal axis are months or number of payments
    months = np.linspace(0, len(balance)-1, len(balance) )
    # plot the salaries for next 20 years
    #plt.figure(figsize=(15,10), dpi=100, facecolor='gray')
    plt.plot(months, balance, 'ro')
    plt.xlabel('Months/Payments')        # add title to the x-axis
    plt.ylabel('Remaining Balance')      # add title to the y-axis 
    plt.show()
    print("Total paid: $", format(total,',.2f'), ".", sep="")
    
interact(credit_card_payment, interest=(0.0,0.1,0.01), payment=(0.00, 100.00, 5) )
````

### Example: Low Rank-Approximations with PCA/SVD

 - $A$ is the sum of the $r$ rank-one matrices:
$$
    A = U \Sigma V^T = \sum_{j=1}^{r} \sigma_{j}u_{j}v_{j}^{T}
$$
 - For any $k$ with $0 \leq k \leq r$, define
$$
    A = \sum_{j=1}^{k} \sigma_{j}u_{j}v_{j}^{T}
$$
Let $k = min(m,n)$, then

$$
    ||A - A_{v}||_{2} = \text{inf}_{B \in \mathbb{C}^{m \times n}} \text{rank}(B)\leq k|| A-B||_{2} = \sigma_{k+1}
$$

- For any $k$ with $0 \leq k \leq r$, the matrix $A_{k}$ also satisfies
$$
    ||A - A_{v}||_{F} = \text{inf}_{B \in \mathbb{C}^{m \times n}} \text{rank}(B)\leq v ||A-B||_{F} = \sqrt{\sigma_{v+1}^{2} + ... + \sigma_{r}^{2}}
$$
